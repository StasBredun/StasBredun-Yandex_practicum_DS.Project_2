Другой столбец с пропущенными значениями — `total_income` — хранит данные о доходах. На сумму дохода сильнее всего влияет тип занятости, поэтому заполнить пропуски в этом столбце нужно медианным значением по каждому типу из столбца `income_type`. ### Категоризация данных

**Задание 16. На основании диапазонов, указанных ниже, создайте в датафрейме `data` столбец `total_income_category` с категориями:**

- 0–30000 — `'E'`;
- 30001–50000 — `'D'`;
- 50001–200000 — `'C'`;
- 200001–1000000 — `'B'`;
- 1000001 и выше — `'A'`.


**Например, кредитополучателю с доходом 25000 нужно назначить категорию `'E'`, а клиенту, получающему 235000, — `'B'`. Используйте собственную функцию с именем `categorize_income()` и метод `apply()`.**

```python
def categorize_income(income):
    try:
        if 0 <= income <= 30000:
            return 'E'
        elif 30001 <= income <= 50000:
            return 'D'
        elif 50001 <= income <= 200000:
            return 'C'
        elif 200001 <= income <= 1000000:
            return 'B'
        elif income >= 1000001:
            return 'A'
    except:
        pass
```

```python
data['total_income_category'] = data['total_income'].apply(categorize_income)
```

**Задание 18. Создайте функцию, которая на основании данных из столбца `purpose` сформирует новый столбец `purpose_category`, в который войдут следующие категории:**

- `'операции с автомобилем'`,
- `'операции с недвижимостью'`,
- `'проведение свадьбы'`,
- `'получение образования'`.

**Например, если в столбце `purpose` находится подстрока `'на покупку автомобиля'`, то в столбце `purpose_category` должна появиться строка `'операции с автомобилем'`.**

**Используйте собственную функцию с именем `categorize_purpose()` и метод `apply()`. Изучите данные в столбце `purpose` и определите, какие подстроки помогут вам правильно определить категорию.**

```python
def categorize_purpose(row):
    try:
        if 'автом' in row:
            return 'операции с автомобилем'
        elif 'жил' in row or 'недвиж' in row:
            return 'операции с недвижимостью'
        elif 'свад' in row:
            return 'проведение свадьбы'
        elif 'образов' in row:
            return 'получение образования'
    except:
        return 'нет категории'
```

```python
data['purpose_category'] = data['purpose'].apply(categorize_purpose)
```

#### 3.1 Есть ли зависимость между количеством детей и возвратом кредита в срок?

```python
#В задаче принимал что в "debt" хранятся данные по правилу: 1 - была задолженность, 0 - не было.
#так же в учет шли 0 (хотя не известно 0 это не брали или брали но вернули в срок)
#Группируем данные по столбцам "кол-во детей" + подсчет категорий, исползовал value_counts и unstack чтобы получить датафрем на выходе
#Добавление столбца общего количесва в подгруппе "кол-во детей", + расчет % от общего числа
# в "positive" храняться те что вернули в срок / "negative" те что не порядочные
#reset_index для того чтобы избавиться от двойних заголовков
#сортировка готовой выборки под непорядчных заемщиков

grouped_children = data.groupby('children')['debt'].value_counts().unstack()
grouped_children['total'] = data.groupby('children')['debt'].count()
grouped_children[1] = grouped_children[1].fillna(0)

grouped_children['positive'] = grouped_children[0] / grouped_children['total'] *  100
grouped_children['negative'] = grouped_children[1] /grouped_children['total'] * 100
grouped_children = grouped_children.reset_index(inplace= False)


grouped_children = grouped_children.sort_values(by='negative', ascending=False)

grouped_children
```

**Вывод:** несмотря не непропорциональное разделение по кол-вам детей, можно сделать вывод что наиболее порядочные заемщики это люди без детей, к выводу не брал бы в учет когда 5 детей т.к таких всего 9 человек. Можно сделать вывод что с увеличением кол-ва детей, расчет % не платежеспособности и они получают негативный рейтинг.

#### 3.2 Есть ли зависимость между семейным положением и возвратом кредита в срок?

```python
#В задаче принимал что в "debt" хранятся данные по правилу: 1 - была задолженность, 0 - не было.
#так же в учет шли 0 (хотя не известно 0 это не брали или брали но вернули в срок)
#Группируем данные по столбцам "семейное положение" + подсчет категорий, исползовал value_counts и unstack чтобы получить датафрем на выходе
#Добавление столбца общего количесва в подгруппе "семейное положение", + расчет % от общего числа
# в positive храняться те что вернули в срок / negative те что не порядочные
#reset_index для того чтобы избавиться от двойних заголовков
#сортировка готовой выборки под непорядчных заемщиков

grouped_family = data.groupby('family_status')['debt'].value_counts().unstack()
grouped_family['total'] = data.groupby('family_status')['debt'].count()


grouped_family['positive'] = grouped_family[0] / grouped_family['total'] *  100
grouped_family['negative'] = grouped_family[1] /grouped_family['total'] * 100

grouped_family = grouped_family.reset_index(inplace= False)


grouped_family = grouped_family.sort_values(by='negative', ascending=False)

grouped_family
```

**Вывод:** В данном случае заемщики более пропорционально распределены по группам. Можно сделать вывод что чем больше у человека жизненого опыта тем более ответственно подходят к возврату займов.
